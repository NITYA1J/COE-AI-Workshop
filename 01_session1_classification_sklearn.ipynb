## A Minimal scikit-learn Classification Workflow

This notebook demonstrates:
- train/test splitting;
- logistic regression;
- majority-class baseline;
- confusion matrix;
- accuracy, precision, recall, and F1-score.

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    confusion_matrix
)

df = pd.read_csv("classification_workshop.csv")
df.head()
```

```python
features = [
    "temperature_c",
    "pressure_kpa",
    "speed_mm_s",
    "vibration_mm_s"
]
target = "defect"

X = df[features]
y = df[target]

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    stratify=y,
    random_state=42
)
```

## Majority-class baseline

```python
majority_class = y_train.mode()[0]
baseline_predictions = np.full(len(y_test), majority_class)

print("Baseline accuracy:",
      round(accuracy_score(y_test, baseline_predictions), 3))
print("Baseline recall:",
      round(recall_score(y_test, baseline_predictions, zero_division=0), 3))
```

```python
df.shape
```

## Logistic regression

```python
model = Pipeline([
    ("scale", StandardScaler()),
    ("classifier", LogisticRegression(max_iter=1000))
])

model.fit(X_train, y_train)
predictions = model.predict(X_test)

print("Accuracy:", round(accuracy_score(y_test, predictions), 3))
print("Precision:", round(precision_score(y_test, predictions, zero_division=0), 3))
print("Recall:", round(recall_score(y_test, predictions, zero_division=0), 3))
print("F1-score:", round(f1_score(y_test, predictions, zero_division=0), 3))
```

```python
cm = confusion_matrix(y_test, predictions)

pd.DataFrame(
    cm,
    index=["Actual acceptable", "Actual defect"],
    columns=["Predicted acceptable", "Predicted defect"]
)
```
