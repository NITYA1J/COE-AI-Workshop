## Visualization with Seaborn and Matplotlib

This notebook introduces three plot types used throughout the workshop:
- scatterplots;
- histograms;
- count plots.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("iris_workshop.csv")
df.head()
```

## Scatterplot: relationship between two inputs

```python
sns.scatterplot(
    data=df,
    x="PetalLength_cm",
    y="PetalWidth_cm",
    hue="Species"
)
plt.title("Petal measurements by species")
plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")
plt.show()
```

## Histogram: distribution of one numerical variable

```python
sns.histplot(
    data=df,
    x="PetalLength_cm",
    hue="Species",
    element="step"
)
plt.title("Distribution of petal length")
plt.xlabel("Petal length (cm)")
plt.show()
```

## Count plot: target balance

```python
sns.countplot(
    data=df,
    x="Species"
)
plt.title("Number of samples per species")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```
