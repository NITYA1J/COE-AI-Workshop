## A Minimal scikit-learn Regression Workflow

Workflow:
1. load data;
2. define inputs and target;
3. split data;
4. establish a baseline;
5. train;
6. predict;
7. evaluate.

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

df = pd.read_csv("regression_workshop.csv")
df.head()
```

```python
features = [
    "temperature_c",
    "pressure_kpa",
    "curing_time_min",
    "material_fraction"
]
target = "strength_mpa"

X = df[features]
y = df[target]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

## Baseline

```python
baseline_predictions = np.full(len(y_test), y_train.mean())

print("Baseline MAE:",
      round(mean_absolute_error(y_test, baseline_predictions), 3))
```

## Linear regression

```python
model = LinearRegression()
model.fit(X_train, y_train)

predictions = model.predict(X_test)

print("Model MAE:", round(mean_absolute_error(y_test, predictions), 3))
print("Model R²:", round(r2_score(y_test, predictions), 3))
```
